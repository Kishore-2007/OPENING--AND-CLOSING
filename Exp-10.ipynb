```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def load_img():
    img =np.zeros((600,700))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,text='KISHORE',org=(100,320), fontFace=font,fontScale= 4,color=(255,255,255),thickness=15,lineType=cv2.LINE_AA)
    return img
def display_img(img):
    plt.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
```

```python
erosion1 = cv2.erode(img, kernel, iterations=1)
```

```python
display_img(erosion1)
```

```python
dilation=cv2.dilate(img,kernel,iterations=1)
```

```python
display_img(dilation)
```

```python
white_noise = np.random.randint(low=0, high=3, size=(600, 700), dtype=np.uint8)
```

```python
white_noise = white_noise*255
```

```python
noise_img = white_noise+img
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
```

```python
display_img(opening)
```

```python
closing = cv2.morphologyEx(noise_img, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```
